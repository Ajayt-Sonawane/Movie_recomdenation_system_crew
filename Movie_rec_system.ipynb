```python
# ✅ STEP 2: Imports
import os
from crewai import Agent, Task, Crew
from langchain_groq import ChatGroq

# ✅ STEP 3: API Key Setup (for Colab)
os.environ["GROQ_API_KEY"] = "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"

# ✅ STEP 4: 🎭 Fun Input - Make it Feel Like Bollywood Chat
print("🎬 Hey cine-lover! 👋 Let's find you the perfect movie 🍿")

print("\n🎭 Question 1:")
print("Imagine your life is a scene in a movie... what's the current vibe?")
print("👉 Options: heartbreak 💔 | victorious 🏆 | bored 😴 | romantic ❤️ | stressed 😩 | mysterious 🕵️‍♂️")
user_feeling = input("🎤 Enter your vibe (or go creative!): ")

print("\n✨ Question 2:")
print("Now tell me… what’s your fantasy for tonight’s watch?")
print("👉 Options: time travel ⏳ | space war 🚀 | deep love story 💑 | brain-twist 🧠 | comedy riot 🤣 | serial killer 🔪")
user_interest = input("🎤 What are you craving in a movie?: ")

# ✅ STEP 5: Define the Movie Guru Agent
movie_agent = Agent(
    role="Filmy Friend",
    goal="Give out-of-the-box movie recommendations tailored to any mood or fantasy",
    backstory="You are the ultimate movie guru — from Bollywood to Hollywood, anime to documentaries — you always know what to watch next!",
    verbose=True,
    allow_delegation=False,
    llm=ChatGroq(model="llama3-8b-8192")
)

# ✅ STEP 6: Mood-Based Task
recommendation_task = Task(
    description=f"""
    My friend is currently feeling '{user_feeling}' and wants something like '{user_interest}' to watch.
    Please recommend 5 epic movies that match their current vibe and craving.
    Include genres, why it's a match, and keep the tone casual and friendly.
    """,
    expected_output="A list of 5 perfect movie recommendations with short, fun reasons for each",
    agent=movie_agent
)

# ✅ STEP 7: Create Crew and Run
crew = Crew(
    agents=[movie_agent],
    tasks=[recommendation_task],
    verbose=True
)

# ✅ STEP 8: Get the Recommendations
result = crew.kickoff()  # kickoff = it is used for creating coordination between the agent and task to run the crew,
print("\n🎥 Your Movie Picks Are In!\n")
print(result)
```
